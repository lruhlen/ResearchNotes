```python
masses = data.Zmass.values/mJup
periods = map(Kepler_semi_to_period,data.ZSTA.values,data.Zmass.values,data.SEMI.values)
periods = numpy.array(periods)/secPerDay
delta_radii = map(get_deltaR,data.Path.values)

# Perform the interpolations...
mi = np.linspace(min(masses), max(masses),75)
pi = np.linspace(min(periods), max(periods),75)
M, P = np.meshgrid(mi, pi)
ri = griddata(masses, periods, delta_radii, mi, pi,interp='nn')
```

```python
from matplotlib import cm
plt.rcParams['figure.figsize'] = [10, 10]  # Sets the overall size of the figure
plt.rcParams['axes.labelsize'] = 14  # Sets the fontsize of the axis labels

fig = plt.figure()
ax = fig.gca()

# Label the axes 
ax.set_ylabel('Mass (mJup)')
ax.set_xlabel('Period (days)')
ax.set_xscale('log')
ax.set_yscale('log')

p = ax.pcolor(P, M, ri, cmap = cm.spectral, vmin=-100, vmax=100,alpha=0.5)
ax.plot(periods,masses,'ko',markersize=10)

cb = plt.colorbar(p,shrink=0.75)
cb.set_label('Percent change in radius')
```

Linear interpolation with matplotlib's griddata routine doesn't seem to work.

When I google this error message, the explanation that comes up deals with a floating point error bug in matplotlib's meshgrid or griddata functions.

Some solutions suggest using scipy's griddata function, instead.

First step in trying this is testing that the example code online works:

```python
def func(x, y):
    return x*(1-x)*np.cos(4*np.pi*x) * np.sin(4*np.pi*y**2)**2

grid_x, grid_y = np.mgrid[0:1:100j, 0:1:200j]
points = np.random.rand(1000, 2)
values = func(points[:,0], points[:,1])
from scipy.interpolate import griddata
#grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest')
#grid_z1 = griddata(points, values, (grid_x, grid_y), method='linear')
grid_z2 = griddata(points, values, (grid_x, grid_y), method='cubic')

print shape(grid_x)
print shape(grid_y)
print shape(values)
print shape(points)
```

Let's try the same thing on a bunch of randomly generated 'eccentricity' values, just to test this with a non-flat distribution.

```python
points = np.array([data.Zmass.values/mJup,data.SEMI.values/AU])
points = points.T # Turns out I needed to transpose the points array to get the routine to work.
grid_x, grid_y = np.mgrid[0:10:100j, 0:1:200j]
values = numpy.random.randn(67)
from scipy.interpolate import griddata
grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest')
grid_z1 = griddata(points, values, (grid_x, grid_y), method='linear')
grid_z2 = griddata(points, values, (grid_x, grid_y), method='cubic')
```

```python
plt.rcParams['figure.figsize'] = [4, 4]  # Sets the overall size of the figure
fig = plt.figure()
ax = fig.gca()
p = ax.pcolor(grid_x, grid_y, grid_z0, cmap = cm.spectral, vmin=-2, vmax=2,alpha=0.5)
plt.colorbar(p,shrink=0.75)
ax.plot(points[:,0],points[:,1],'ko',markersize=5)
```

```python
plt.rcParams['figure.figsize'] = [4, 4]  # Sets the overall size of the figure
fig = plt.figure()
ax = fig.gca()
p = ax.pcolor(grid_x, grid_y, grid_z1, cmap = cm.spectral, vmin=-2, vmax=2,alpha=0.5)
plt.colorbar(p,shrink=0.75)
ax.plot(points[:,0],points[:,1],'ko',markersize=5)
```

```python
plt.rcParams['figure.figsize'] = [4, 4]  # Sets the overall size of the figure
fig = plt.figure()
ax = fig.gca()
p = ax.pcolor(grid_x, grid_y, grid_z2, cmap = cm.spectral, vmin=-2, vmax=2,alpha=0.5)
plt.colorbar(p,shrink=0.75)
ax.plot(points[:,0],points[:,1],'ko',markersize=5)
```
