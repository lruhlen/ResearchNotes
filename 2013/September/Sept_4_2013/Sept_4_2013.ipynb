##Adding arrows to the dataA regular point-and-line plot is okay, but it doesn't tell us which direction the planets' radii have evolved in.It would be nice to add arrows to the data to show this, but first I need to figure out how to plot 3D arrows.[Someone already did this.](http://stackoverflow.com/questions/11140163/python-matplotlib-plotting-a-3d-cube-a-sphere-and-a-vector)  Here's their code, slightly modified/tinkered with.

```python
from matplotlib.patches import FancyArrowPatchfrom mpl_toolkits.mplot3d import proj3dfig = plt.figure()ax = fig.gca(projection='3d')class Arrow3D(FancyArrowPatch):    def __init__(self, xs, ys, zs, *args, **kwargs):        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)        self._verts3d = xs, ys, zs    def draw(self, renderer):        xs3d, ys3d, zs3d = self._verts3d        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))        FancyArrowPatch.draw(self, renderer)a = Arrow3D([0,2],[-1,3],[4,-2], mutation_scale=30, lw=1, arrowstyle="-|>", color="magenta")ax.add_artist(a)ax.set_xlim([-4,4])ax.set_ylim([-4,4])ax.set_zlim([-4,4])a = Arrow3D([4,-3],[4,-3],[3,-4], mutation_scale=30, lw=5, arrowstyle="-|>", color="c")ax.add_artist(a)plt.show()
```
